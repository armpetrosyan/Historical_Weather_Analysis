```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('daily_weather_data.csv')
first_last_5 = pd.concat([df.head(), df.tail()])
first_last_5
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
(df.drop_duplicates(inplace=True))
```

```python
print(round(df.describe(),2))
```

```python
df.info()
```

```python
plt.figure(figsize=(30, 6))
plt.plot(df.index, df['tavg'])
plt.xlabel('Indexs')
plt.ylabel('Temperature')
plt.title('Temperature Trends Over Time')
plt.show()
```

```python
plt.scatter(df['Longitude'], df['Latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('City Locations')
plt.show()
```

```python
plt.hist(df['tavg'], bins=30)
plt.xlabel('Average Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Distribution')
plt.show()
```

```python
plt.figure(figsize=(125, 20))
sns.barplot(x='country', y='tavg', data=df)
plt.xlabel('Country')
plt.ylabel('Average Temperature')
plt.title('Average Temperature per Country')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(100, 6))

sns.violinplot(x='country', y='wspd', data=df)
plt.xlabel('Country')
plt.ylabel('Wind Speed')
plt.title('Wind Speed Distribution per Country')

plt.xticks(rotation=45, fontsize=8)
plt.tight_layout()
plt.show()
```
